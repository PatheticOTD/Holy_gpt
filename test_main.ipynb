```python
import pandas as pd
import nltk
import os
import numpy as np
from math import log
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("karkavelrajaj/amazon-sales-dataset")
filename = os.listdir(path)[0]
print("Path to dataset files:", path)
```

```python
nltk.download('punkt')
```

```python
df = pd.read_csv(os.path.join(path, filename))
```

```python
sentenses = ' '.join(df['review_content'][:500].values)
st = nltk.tokenize.PunktSentenceTokenizer()
words = nltk.tokenize.word_tokenize(sentenses)
separated_sentenses = st.tokenize(sentenses)

print(len(separated_sentenses), len(words))
```

```python
vocab = {}

for i in set(words):
    for j in words:
        if i == j:
            vocab[i] = vocab.get(i, 0) + 1
```

```python
vocab['Hello']
```

```python
dick = {}

for i in range(1, len(words)):
    probs =  (dick.get(f"{words[i-1]} {words[i]}", 0) + 1)
    dick[f"{words[i-1]} {words[i]}"] = probs
for key in dick.keys():
    logit = log(dick[key] / vocab[key.split()[0]])
    dick[key] = -logit if logit < 0 else logit
    
```

```python
dick
```

```python
n = 100
starting_word = 'LG'
ans = [starting_word]
curr_word = starting_word
for i in range(n):
    max_proba = 0
    word = ''
    for j in dick:
        if j.split()[0] == ans[-1]:
            if dick[j] > max_proba:
                max_proba = dick[j]
                word = j.split()[1]
    ans.append(word)
    if max_proba == 0:
        break
' '.join(ans)
```
